# Define the Model 
Build CNN with
- Convolutional Layer 1: 32 filters, (3x3) kernel, ReLU 
- MaxPooling Layer 1: (2x2) pool size
- Convolutional Layer 2: 64 filters, (3x3) kernel, ReLU
- MaxPooling Layer 2: (2x2) pool size
- Flaten Layer
- Dense Layer: 128 units, ReLU 
- Output Layer: 10 units (for 10 classes), softmax
- Compile the baseline model with appropriate loss and metrics

```python
model_1CNN = tf.keras.models.Sequential([
tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
tf.keras.layers.MaxPooling2D(2, 2),
tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(128, activation='relu'),
tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
#use this callback to monitor the training time
from timeit import default_timer as timer

class TimingCallback(keras.callbacks.Callback):
    def __init__(self, logs={}):
        self.logs=[]
    def on_epoch_begin(self, epoch, logs={}):
        self.starttime = timer()
    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(timer()-self.starttime)

cb = TimingCallback()

```

# Model 1
## Run non-augmented from loaded memory

For all the models lets use a batch size of 32 and 15 epochs

```python
model_1CNN.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model_1CNN.fit(train_images, train_enc, epochs=15, batch_size=32, validation_split=0.2, callbacks=[cb])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
