# Define the Model 
Build CNN with
- Convolutional Layer 1: 32 filters, (3x3) kernel, ReLU 
- MaxPooling Layer 1: (2x2) pool size
- Convolutional Layer 2: 64 filters, (3x3) kernel, ReLU
- MaxPooling Layer 2: (2x2) pool size
- Flaten Layer
- Dense Layer: 128 units, ReLU 
- Output Layer: 10 units (for 10 classes), softmax
- Compile the baseline model with appropriate loss and metrics


For all the models lets use a batch size of 32, 15 epochs, adam optimiser, and categorical crossentropy

```python
model_1CNN = tf.keras.models.Sequential([
tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
tf.keras.layers.MaxPooling2D(2, 2),
tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(128, activation='relu'),
tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
#use this callback to monitor the training time
from timeit import default_timer as timer

class TimingCallback(keras.callbacks.Callback):
    def __init__(self, logs={}):
        self.logs=[]
    def on_epoch_begin(self, epoch, logs={}):
        self.starttime = timer()
    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(timer()-self.starttime)

cb = TimingCallback()

```

# Analysis and Discussion

        Both models utalizing memory trained significatly faster than those pulling from the disk. However, there will be instances in which not all of the data can be loaded into memory, and thus it is importatnt to understand both.The addition of augmentation also significantly increased training time, most likely due to the extra process after the image is aquired. 
         While the augmented from memory was clearly malfunctioning at the beginning, eventually it was corrected and demonstrates that fact that data augmentation greatly reduced the overfitting as both validation and training followed the same trend. In the non-augmented data there is a visible drift where the validation begins to taper off and the training accuracy continues to improve, demonstrating overfitting. By changing the images slightly, we are ensuring that the model is not memorizing the data and overfitting, but continuing to test and learn the actual patterns within the image.
         Accuracy was higher in the 15 epoch training sessions for the non-augmented data, but the trend in loss and accuracy of the augmented data models indicate that more epochs may result in higher accuracy. However, with model training there must be a tradeoff between accuracy, memory, and time.
